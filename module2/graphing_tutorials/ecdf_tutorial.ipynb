```python
from matplotlib import pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
sns.ecdfplot(x="tip", data=tips);
```

```python
sns.ecdfplot(x="tip", data=tips);
plt.axvline(4, c="black");
```

```python
sns.ecdfplot(x="tip", data=tips, hue="day");
```

### Investigate joining on country

Before we can proceed, we need to make sure that the columns used to join the data--namely the country--actually match.  Do this by

1. For each table, select just the country columns and make sure the column names match.
2. In each table add a file column that contains the respective file name
3. Use distinct on each table to remove duplicates
4. Perform a full outer join and filter on rows that didn't match (i.e. with a missing value in either column).
5. Sort the table  by the country name column and use distinct again.
6. Write the resulting table to a file and open it.

```python
covid_countries = (covid_data_cleaned
                       >> select(X["Country/Region"])
                       >> rename(country = "Country/Region")
                       >> mutate(file = "Covid")
                       >> distinct
                  )
covid_countries.head()
```

```python
wb_countries = (wb_data_cleaned
                    >> select(X["Country Name"])
                    >> rename(country = "Country Name")
                    >> mutate(file = "World Bank")
                    >> distinct
               )
wb_countries.head()
```

```python
mismatched_countries = (covid_countries
                            >> outer_join(wb_countries, by="country")
                            >> filter_by(X.file_x.isna() | X.file_y.isna())
                            >> arrange(X.country)
                            >> distinct # don't think this actually does anything since we did distinct on the individual tables already
                        )
mismatched_countries.head()
```

```python
mismatched_countries.to_csv("./data/mismatched_countries.csv")
```
